# For T-testing, let's compare the cholesterol levels (chol) between two groups, 
# such as those with and without heart disease (target).

```python
import pandas as pd
from scipy import stats

# Load the dataset
heart_data = pd.read_csv("C:/Users/ankuv/Downloads/Heart Disease data.csv")

# 1. One-Sample T-Test
# Hypothesis: H0: The mean cholesterol level is equal to 200, H1: The mean cholesterol level is not equal to 200
cholesterol_mean = 200

# Perform the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(heart_data['chol'], cholesterol_mean)

print("One-Sample T-Test")
print(f"Hypotheses:")
print(f"H0: The mean cholesterol level is 200.")
print(f"H1: The mean cholesterol level is not 200.\n")

print(f"Results:")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis (H0). The mean cholesterol level is significantly different from 200.\n")
else:
    print("Conclusion: Fail to reject the null hypothesis (H0). No significant difference from 200.\n")


# 2. Two-Sample T-Test
# Hypothesis: H0: The mean cholesterol levels are equal between the two groups, H1: The mean cholesterol levels are not equal between the two groups

# Separate the data into two groups based on the target variable
chol_with_disease = heart_data[heart_data['target'] == 1]['chol']
chol_without_disease = heart_data[heart_data['target'] == 0]['chol']

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(chol_with_disease, chol_without_disease)

print("Two-Sample T-Test")
print(f"Hypotheses:")
print(f"H0: The mean cholesterol levels are equal between the two groups (with and without heart disease).")
print(f"H1: The mean cholesterol levels are not equal between the two groups.\n")

print(f"Results:")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis (H0). There is a significant difference in mean cholesterol levels between the two groups.\n")
else:
    print("Conclusion: Fail to reject the null hypothesis (H0). No significant difference in mean cholesterol levels between the two groups.\n")
```
